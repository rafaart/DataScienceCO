```python
def fetch_bcb(codigo, data_inicio, data_fim):
  url = f'https://api.bcb.gov.br/dados/serie/bcdata.sgs.{codigo}/dados?formato=json&dataInicial={data_inicio}&dataFinal={data_fim}'
  df = pd.read_json(url)
  return df
```
